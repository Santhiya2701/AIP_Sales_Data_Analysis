```python
import matplotlib.pyplot as plt

product_sales = df.groupby("PRODUCTLINE")["SALES"].sum()

product_sales.plot(kind="bar", title="Sales by Product Line")
plt.show()
```

INSIGHTS:
- Some product lines generate higher sales than others.
- Classic Cars and Vintage Cars perform well.
- Sales are not evenly distributed across categories.


```python
# Convert ORDERDATE to datetime
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Extract month
df['MONTH'] = df['ORDERDATE'].dt.month

# Monthly sales
monthly_sales = df.groupby('MONTH')['SALES'].sum()

# Plot
monthly_sales.plot(kind='line', title='Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

INSIGHTS:
- Sales vary significantly across different months.
- Towards the end of the year, sales increase sharply.
- This indicates seasonal demand, possibly due to festivals or year-end purchases.
